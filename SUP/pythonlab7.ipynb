```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
np.random.seed(1)
#give same when you excute again and again 
data=5*np.random.randn(100)+50  #creating 100 samples
print((np.mean(data)),np.std(data))
```

```python
#histogram also one type of normaality check


plt.hist(data,bins=10,color='orange')
plt.show()
```

```python
#qq plots

from statsmodels.graphics.gofplots import qqplot
data=5*np.random.randn(50)+50
qqplot(data,line='s',color='red')
plt.show()
```

```python
#we need to  draw a normally disstributed data using function norm
from scipy.stats import norm  #normally distributed data
data=norm.rvs(loc=0,scale=1,size=1000)
qqplot(data,line='s')
```

```python
#probability plot

data=norm.rvs(loc=0,scale=1,size=1000)
stats.probplot(data,plot=plt)
```

```python
#statstical tests
#by using this we can tell about exact measure
#it will generate two value i.e statics value and p value
#pvalue===whether it should be accepted as normal distribution or not
#if upto 5% deviation  it should be accepted 
#p<=alpha.rejectH0   accept data otherwise reject the data
```

```python
#shapiro will test or ws test
#random nummbers also fall under norml distribution
from numpy.random import seed
from numpy.random import randn
from scipy.stats import  shapiro
seed(1)
#generate univarte data
data=5*randn(100)+50

#normality ceck
stat,p=shapiro(data)
print('Statics=%.3f,p=%.3f'%(stat,p))
alpha=0.05
if p>alpha:
    print("Sample looks Gaussian Distribution(fail to reject H0)")
else:
    print("Sample doesnot looks Gaussian Distribution(rject H0)")
```

```python
#anderson daarling test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import  anderson
seed(1)
#generate univarte data
data=5*randn(100)+50
result=anderson(data)
print("Statstics: %.3f"%result.statstics)
```

```python
#D'Agostains pearson k^2 test
from scipy.stats import normaltest
stat,p=normaltest(data)
print('Statics=%.3f,p=%.3f'%(stat,p))
alpha=0.05
if p>alpha:
    print("Sample looks Gaussian Distribution(fail to reject H0)")
else:
    print("Sample doesnot looks Gaussian Distribution(rject H0)")
```
